# Visualize the performed benchmarks

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Save figures as png? 
save_figs = False
# Define image name (for export)
FIGNAME = "Benchmark_12_23"

## Define Filenames 

# with native python code
FILENAME_py = "results/Py_Benchmark_Results.csv" 
# with numba
FILENAME_nu = "results/numba_Benchmark_Results.csv" 
# with transonic
FILENAME_tr = "results/transonic_Benchmark_Results.csv" 
```

```python
# import python result values
df_py = pd.read_csv(FILENAME_py)

# import numba result values
df_nu = pd.read_csv(FILENAME_nu)

# import transonic result values
df_tr = pd.read_csv(FILENAME_tr)
#df_tr
```

# Plot results 

```python
## Plot time vs elems for each alg and python approach
plt.semilogy(df_py['N'],df_py['t_G1'], df_py['N'],df_py['t_R1'], df_py['N'],df_py['t_FRLT'])

plt.title('Time VS No. of Elem \nfor different alg by python approach')
plt.legend(['G1 Python', 'R1 Python', 'FRTL Python'])

plt.xlabel('No. of Elems [-]')
plt.ylabel('Time [s]')
plt.grid(visible=True)
if save_figs:
    plt.savefig(''.join(['images/',FIGNAME,'_time_python.png']), dpi=600)
plt.show()
```

```python
## Plot time vs elems for each alg and numba approach
plt.semilogy(df_nu['N'],df_nu['t_G1'], df_nu['N'],df_nu['t_R1'], df_nu['N'],df_nu['t_FRLT'])

plt.title('Time VS No. of Elem \nfor different alg by numba approach')
plt.legend(['G1 Numba', 'R1 Numba', 'FRTL Numba'])

plt.xlabel('No. of Elems [-]')
plt.ylabel('Time [s]')
plt.grid(visible=True)
if save_figs:
    plt.savefig(''.join(['images/',FIGNAME,'_time_numby.png']), dpi=600)
plt.show()
```

```python
## Plot time vs elems for each alg and transonic approach
plt.semilogy(df_tr['N'],df_tr['t_G1'], df_tr['N'],df_tr['t_R1'], df_tr['N'],df_tr['t_FRLT'])

plt.title('Time VS No. of Elem \nfor different alg by transonic approach')
plt.legend(['G1 Transonic', 'R1 Transonic', 'FRTL Transonic'])

plt.xlabel('No. of Elems [-]')
plt.ylabel('Time [s]')
plt.grid(visible=True)
if save_figs:
    plt.savefig(''.join(['images/',FIGNAME,'_time_transonic.png']), dpi=600)
plt.show()
```

```python
## Plot absolute error vs elems for each alg and each approach
plt.loglog(df_py['N'],df_py['e_G1_max'], df_nu['N'],df_nu['e_G1_max'], df_tr['N'],df_tr['e_G1_max'],
         df_py['N'],df_py['e_R1_max'], df_nu['N'],df_nu['e_R1_max'], df_tr['N'],df_tr['e_R1_max'],
         df_py['N'],df_py['e_FRLT_max'], df_nu['N'],df_nu['e_FRLT_max'], df_tr['N'],df_tr['e_FRLT_max'])

plt.title('Max absolute error vs no. of elem \nfor different alg by different approaches')
plt.legend(['G1 Python', 'G1 Numba', 'G1 Transonic',
            'R1 Python', 'R1 Numba', 'R1 Transonic',
            'FRTL Python', 'FRTL Numba', 'FRTL Transonic'])
plt.xlabel('No. of Elems [-]')
plt.ylabel('Absolute Error [-]')
plt.grid(visible=True)
if save_figs:
    plt.savefig(''.join(['images/',FIGNAME,'_abs_err.png']), dpi=600)
plt.show()
```

```python
## Plot absolute error vs elems for each alg and each approach
plt.loglog(df_py['N'],df_py['e_rel_G1_max'], df_nu['N'],df_nu['e_rel_G1_max'], df_tr['N'],df_tr['e_rel_G1_max'],
         df_py['N'],df_py['e_rel_FRLT_max'], df_nu['N'],df_nu['e_rel_FRLT_max'], df_tr['N'],df_tr['e_rel_FRLT_max'])

plt.title('Max relative error vs no. of elem \nfor different alg by different approaches')
plt.legend(['G1 Python', 'G1 Numba', 'G1 Transonic',
            'FRTL Python', 'FRTL Numba', 'FRTL Transonic'])
plt.xlabel('No. of Elems [-]')
plt.ylabel('Relative Error [-]')
plt.grid(visible=True)
if save_figs:
    plt.savefig(''.join(['images/',FIGNAME,'_rel_err_1.png']), dpi=600)
plt.show()
```

```python
## Plot absolute error vs elems for each alg and each approach
plt.loglog(df_py['N'],df_py['e_rel_R1_max'], df_nu['N'],df_nu['e_rel_R1_max'], df_tr['N'],df_tr['e_rel_R1_max'])

plt.title('Max relative error vs no. of elem \nfor different alg by different approaches')
plt.legend(['R1 Python', 'R1 Numba', 'R1 Transonic'])
plt.xlabel('No. of Elems [-]')
plt.ylabel('Relative Error [-]')
plt.grid(visible=True)
if save_figs:
    plt.savefig(''.join(['images/',FIGNAME,'_rel_err_2.png']), dpi=600)
plt.show()
```
